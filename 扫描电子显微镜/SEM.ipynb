```python
# -*- coding: utf-8 -*-
"""
Created on Wed Mar 13 23:39:32 2024

@author: DELL
"""


import numpy as np
from PIL import Image
import imageio
import matplotlib.pyplot as plt
filename = "C:\\Users\\DELL\\Downloads\\retest.jpg"

img = imageio.v3.imread(filename,pilmode='RGB')

h,w,c = img.shape

print(h,w,c)
pixels = np.array(img)
print(pixels.shape)
        
  
```

```python
red = pixels[:,:,0]
print(red.shape)
```

```python
Z=np.fft.fft(red)
print(Z.shape)
```

```python
def show(Z):
    height,width = Z.shape
    img1 = Image.new('RGB', (width, height), (255, 255, 255))

    for x in range(height):
        for y in range(width):
            red = int(Z[x][y].real)
            blue = int(Z[x][y].imag)
            green = int((red+blue)/2)
            img1.putpixel((y, x), (red, green, blue))
    img1.show()
```

```python
show(Z)
```

```python
def Plot(X):
    x = [i for i in range(len(X))]
    plt.plot(x,X)
    plt.xlabel('freq:cycs/1024pixel')
    plt.ylabel('spectrum density in log10')
    plt.show()
    
```

```python
ZZ = (Z*np.conjugate(Z)).real
```

```python
freq = np.mean(ZZ, axis = 0)
```

```python
Plot(np.log10(freq[1:300]+0.0005))
```
